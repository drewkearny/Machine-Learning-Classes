# 7. Fizz Buzz

For each number in 1,2,...,50
 1. print only the word "fizz" if the number is divisible by 3
 2. print only the word "buzz" if the number if divisible by 5
 3. print only the word "fizzbuzz" if the number is divisible by 3 and 5
 4. If the number is not divisible by any of these, just print the number.

hint: We can use the `range()` function to simplify our looping range. Check divisibility with the modulo operator %.

> Indented block



```python
# hint: Modulo operator: a % b. if b divides a, then a % b == 0
print(10 % 3)
print(10 % 2)
```

```python
# your code here
# I changed the end range in the function to 51 to include 50 in the rande

for n in range(1, 51):
    if n % 3 == 0 and n % 5 == 0:
      print("fizzbuzz")
    elif n % 3 == 0:
        print("fizz")
    elif n % 5 == 0:
        print("buzz")
    else:
        print(n)
```
